```python
traj['Rmag'] = traj['Rx(km)[J2000-EARTH]']**2 + traj['Ry(km)[J2000-EARTH]']**2 + traj['Rz(km)[J2000-EARTH]']**2
traj['Rmag'] = np.sqrt(traj['Rmag'])
```

```python
traj
```

```python
plt.plot(traj['MISSION ELAPSED TIME (mins)'],traj['Rmag'])
plt.plot(traj['MISSION ELAPSED TIME (mins)'],traj['Rx(km)[J2000-EARTH]'])
plt.plot(traj['MISSION ELAPSED TIME (mins)'],traj['Ry(km)[J2000-EARTH]'])
plt.plot(traj['MISSION ELAPSED TIME (mins)'],traj['Rz(km)[J2000-EARTH]'])
ax = plt.gca()
ax.set_xlabel('Time (mins)')
ax.set_ylabel('Range (km)')
min(traj['Rmag'])
```

```python
traj['Vmag'] = traj['Vx(km/s)[J2000-EARTH]']**2 + traj['Vy(km/s)[J2000-EARTH]']**2 + traj['Vz(km/s)[J2000-EARTH]']**2
traj['Vmag'] = np.sqrt(traj['Vmag'])
```

```python
plt.plot(traj['MISSION ELAPSED TIME (mins)'],traj['Vmag'])
#plt.plot(traj['MISSION ELAPSED TIME (mins)'],traj['Vx(km/s)[J2000-EARTH]'])
#plt.plot(traj['MISSION ELAPSED TIME (mins)'],traj['Vy(km/s)[J2000-EARTH]'])
#plt.plot(traj['MISSION ELAPSED TIME (mins)'],traj['Vz(km/s)[J2000-EARTH]'])
ax = plt.gca()
ax.set_xlabel('Time (mins)')
ax.set_ylabel('Speed (m/s)')
min(traj['Vmag'])
```
